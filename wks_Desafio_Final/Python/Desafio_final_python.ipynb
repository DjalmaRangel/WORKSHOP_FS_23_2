### **CARREGAMENTO DOS DADOS**

```python
# Carregamento e visualização de uma amostra dos dados com o intuito de conhecer as colunas e os conteúdos de cada uma delas
df = pd.read_csv('/content/dados_rh.csv')
df.head()
```

### **AGRUPANDO OS DADOS**

#### Agrupamento 1 - Número de Trabalhadores por departamento

```python
# Criando novo dataframe com o agrupamento do Número de trabalhadores por Departamento
df_grupo1 = df.groupby('Department')[['EmployeeCount']].count()
df_grupo1.sort_values('Department')
```

```python
# Criando gráfico de barras informando ao Número de trabalhadores por Departamento
df_grupo1.plot(kind = 'bar')
```

Observa-se que o Departamento de Pesquisa e Desenvolvimento (Research & Development) possui muito mais funcionários do que os outros departamentos.

#### Agrupamento 2 - Número de funcionários por Nível de Escolaridade

```python
# Criando novo dataframe com o agrupamento do número de funcionários por Nível de Escolaridade
df_grupo2 = df.groupby('Education')[['EmployeeCount']].count()
df_grupo2.sort_values('Education')
```

```python
# Criando gráfico de linha informando o número de funcionários por Nível de Escolaridade
df_grupo2.plot(kind = 'line', marker = 'o')
```

Observa-se que maioria dos trabalhadores tem "*nível de escolaridade 3*" e minoria tem "*nível de escolaridade 5*".

#### Agrupamento 3 - Número de Funcionários por Incidência de Atrito na empresa

```python
# Criando novo dataframe com o agrupamento do número de funcionários que registraram ter  de atrito na empresa
df_grupo3 = df.groupby('Attrition')[['EmployeeCount']].count()
df_grupo3.sort_values('Attrition')
```

```python
df_grupo3.plot(kind = 'pie', subplots=True)
```

Observa-se que a grande maioria não teve Atrito na empresa.
